# Vote share of each party at state level

```
## all unique parties in the state
party_names=df['party_name'].unique()

vote_share=(df.groupby(by='party_name').sum()["total_valid"]/sum(df["total_valid"])).reset_index()

votes=df.groupby(by='party_name').sum()["total_valid"].reset_index()
# vote_output=pd.concat([party_names,vote_share,votes],axis=1)
```

```
vote_output=pd.merge(vote_share,votes,on='party_name')
vote_output.columns=["party_name","valid_vote_share","total_votes"]
vote_output
```

```
## get the alliance of each party
def get_alliance(party_name):
    alliance_name=""
    party_name=party_name.upper()
    for alliance, parties in alliance_dict_16.items():
        for p in parties.values():
            if party_name in p:
                alliance_name=alliance

    if alliance_name=="":
        alliance_name="Others"

    return alliance_name

vote_alliance=vote_output.party_name.apply(get_alliance)
vote_alliance
```

```
vote_output=pd.concat([vote_output,vote_alliance],axis=1)
vote_output.columns=["party_name","valid_vote_share","total_votes","party_alliance"]
vote_output
```

```
vote_output.to_csv("/content/drive/My Drive/Colab Notebooks/India_Election_Analytics/vote_shares_state_level_16.csv",index=False)
```

# Vote shares of all parties in each constituency, sorted by the total votes in descending order

```
df_temp=df[['ac_name','party_name','total_valid']]
```

```
df_temp['vote_share']=df_temp.groupby(by='ac_name').total_valid.apply(lambda x: x/x.sum())
df_temp
```

```
df_party_votes_ac_level = df_temp.set_index(['ac_name',df.groupby(['ac_no','ac_name']).cumcount()+1]).unstack().sort_index(level=1, axis=1)
df_party_votes_ac_level.columns = df_party_votes_ac_level.columns.map('{0[0]}_{0[1]}'.format)
df_party_votes_ac_level=df_party_votes_ac_level.reset_index()
df_party_votes_ac_level
```

```
df_party_votes_ac_level.to_csv("/content/drive/My Drive/Colab Notebooks/India_Election_Analytics/all_party_vote_shares_ac_level_16.csv",index=False)
```

# Vote shares of 10 biggest parties in each constituency

```
## 10 biggest party names except "IND" and "NOTA"
vote_output.loc[(vote_output.party_name!="IND") & (vote_output.party_name!="NOTA"),]\
                .sort_values(\
                    by='valid_vote_share',\
                    ascending=False)\
                .reset_index(drop=True)\
                .loc[:11,]
```

```
first_12_parties=vote_output.loc[(vote_output.party_name!="IND") & (vote_output.party_name!="NOTA"),]\
                .sort_values(\
                    by='valid_vote_share',\
                    ascending=False)\
                .reset_index(drop=True)\
                .loc[:11,].party_name
first_12_parties=list(first_12_parties)
first_12_parties
```

```
vote_share_12_parties=pd.DataFrame(columns=["ac_name"]+first_12_parties)

## ac_name
vote_share_12_parties['ac_name']=df_temp['ac_name'].unique()

## vote share
# df_party_votes_ac_level.fillna(0)
for ac_no in range(len(df_party_votes_ac_level)):
    ac_name=df_party_votes_ac_level.loc[ac_no,'ac_name']
    
    for party in first_12_parties:
        if df_party_votes_ac_level.loc[ac_no,].str.contains(party).sum()==0:
            vote_share_12_parties.loc[ac_no,party]=0
        else:
            index=list(df_party_votes_ac_level.loc[ac_no,].str.contains(party)).index(True)
            vote_share_12_parties.loc[ac_no,party]=df_party_votes_ac_level.iloc[ac_no,index+2]

vote_share_12_parties
```
