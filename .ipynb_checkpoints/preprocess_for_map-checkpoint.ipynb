```python
import numpy as np
import pandas as pd
```

```python
file="/content/drive/My Drive/Colab Notebooks/India_Election_Analytics/polling stations-odd/2016-odd.xlsx"
directory="/content/drive/My Drive/Colab Notebooks/India_Election_Analytics/polling stations-odd/ally_output/"
```

```python
alliance_names=["AIADMK+","DMK","PWF","NDA","NTK","PMK","Others"] ## unique alliances 

aiadmk_parties=["All India Anna Dravida Munnetra Kazhagam","ADMK","AIADMK",
                "Manithaneya Jananayaga Katchi","MJK",
                "Republican Party of India","RPI",
                "Samathuva Makkal Katchi","SMK","AISMK",
                "Tamil Nadu Kongu Ilaingar Peravai","TNKIP",
                "Mukkulathor Pulipadai","MP",
                "Tamil Maanila Muslim League","TMML"]
aiadmk_upper=[x.upper() for x in aiadmk_parties]

dmk_parties=["Dravida Munnetra Kazhagam","DMK",
             "Indian National Congress","INC",
             "Indian Union Muslim League","IUML",
             "Manithaneya Makkal Katchi","MMK","MAMAK",
             "Puthiya Tamilakam","PT",
             "Perunthalaivar Makkal Katchi","PMK",
             "Tamil Nadu Peasants and Workers Party","TNPWP",
             "Samuga Samathuva Padai","SSP"]
dmk_upper=[x.upper() for x in dmk_parties]

pwf_parties=["Desiya Murpokku Dravida Kazhagam","DMDK",
             "Marumalarchi Dravida Munnetra Kazhagam","MDMK",
             "Communist Party of India",\
             "Communist Party Of India Tamilaga Makkal Munnetra \
             Kazhagam Pattali Makkal Katchi Naam Tamilar Katchi","CPI",
             "Communist Party of India (Marxist)","CPI(M)","CPIM","CPI (M)",
             "Tamil Manila Congress","TMC","TMC(M)","TMC (M)",
             "Viduthalai Chiruthaigal Katchi","VCK"]
pwf_upper=[x.upper() for x in pwf_parties]


nda_parties=["Bharatiya Janata Party","BJP",
             "Indhiya Jananayaga Katchi","IJK"]
nda_upper=[x.upper() for x in nda_parties]

pmk_party=["Pattali Makkal Katchi","PMK" #independent]
pmk_upper=[x.upper() for x in pmk_parties]  
           
ntk_party=["Naam Tamilar Katchi","NTK", #independent]
ntk_upper=[x.upper() for x in ntk_parties]  

other_parties=["Tamilnadu Youth Party","TNYP",
        "KMDK",
        "NONE OF THE ABOVE","NOTA","SDPI",
        #...
        "Independent","INDI"]
other_upper=[x.upper() for x in other_parties]

alliance_dict=dict(zip(alliance_names,
                  [aiadmk_upper,dmk_upper,pwf_upper,nda_upper,other_upper]))
```

```python
def series_upper(series):
    ## change all strings to upper
    return series.str.upper()

def read_data(n): 
    ## input is constituency no.
    ## have to change it into the real no in the table
    ## starting from 0 sheet
    real_no=int((n-1)/2)
    data=pd.read_excel(file,sheet_name=real_no)
    data.fillna("None",inplace=True)
    return data

def get_alliance(data):
    alliances={}
    count=0
    for party in data.loc[1,data.loc[1,]!="None"].str.upper():
        count+=1
        for alliance, parties in alliance_dict.items():
            if party in parties:
                alliances[count]=alliance
        if count not in alliances.keys():
            alliances[count]="Others"
    alliances_=list(alliances.values())
    return alliances_

def get_ally_votes(data):
    alliances=get_alliance(data)

    temp_df=pd.DataFrame()
    temp_df=data.loc[2:len(data)-4,data.loc[1,]!="None"]
    temp_df.columns=alliances
    temp_df.reset_index(inplace=True,drop=True)

    output=pd.DataFrame()
    for name in alliance_names:
        temp=temp_df.loc[:,temp_df.columns==name]
        output[name]=temp.apply(sum,axis=1)

    ## total
    output.loc["Total"]=output.apply(sum)
    output.fillna(0,inplace=True)
    return output
```

```python
odd_const=[(2*x+1) for x in range(117)]

for constituency in odd_const:
    print(constituency)
    data1=read_data(constituency)
    output_df=get_ally_votes(data1)
    output_df.to_csv(f"{directory}ally_{constituency}.csv")
```
